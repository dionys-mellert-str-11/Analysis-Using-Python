```python
#load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
```

```python
#look at excel file at certain folder, pull csv file and change to png file

os.chdir(r'C:\Users\he1108434\Documents\EPS_original_data\2021-1-2')

fileName = 'Part_maXYmos_NC_MP-006_2021-01-02_23-37-29_2150567_NOK.csv'

with open(fileName) as fp:
    lines = fp.readlines()

index = 279
curveData = [[],[]]

#print(lines)
while lines[index] != '\n':
    datapoint = lines[index].split(';')[1:3]
    datapoint[0] = datapoint[0].replace(',', '.')
    datapoint[1] = datapoint[1].replace(',', '.')
    curveData[0].append(float(datapoint[0]))
    curveData[1].append(float(datapoint[1]))
    index = index + 1

fig, ax = plt.subplots()
ax.plot(curveData[0], curveData[1], linewidth=7.0)
ax.set_ylabel('kN')
ax.set_xlabel('mm')
plt.axis([min(curveData[0]), max(curveData[0]), min(curveData[1]), max(curveData[1])])
fig.set_size_inches(15, 10.5, forward=True)
os.chdir(r'C:\Users\he1108434\Downloads')
#plt.title('test1')
plt.savefig('Part_maXYmos_NC_MP-006_2021-01-02_23-37-29_2150567_NOK')
plt.show()
```
