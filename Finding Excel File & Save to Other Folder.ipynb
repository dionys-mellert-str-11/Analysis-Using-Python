```python
#find a file in Python folder
import matplotlib.pyplot as plt
import os
for root, dirs, files in os.walk(r'C:\Users\he1108434\Documents\EPS_raw_data\2021-2-12'):
    for file in files:
        if file.endswith("NOK.csv"):
            path_file = os.path.join(root,file)
            print(path_file)
            with open(path_file) as fp:
                lines = fp.readlines()

            index = 279
            curveData = [[],[]]

            #print(lines)
            while lines[index] != '\n':
                datapoint = lines[index].split(';')[1:3]
                datapoint[0] = datapoint[0].replace(',', '.')
                datapoint[1] = datapoint[1].replace(',', '.')
                curveData[0].append(float(datapoint[0]))
                curveData[1].append(float(datapoint[1]))
                index = index + 1

            fig, ax = plt.subplots()
            ax.plot(curveData[0], curveData[1], linewidth=7.0)
            ax.set_ylabel('kN')
            ax.set_xlabel('mm')
            plt.axis([min(curveData[0]), max(curveData[0]), min(curveData[1]), max(curveData[1])])
            fig.set_size_inches(15, 10.5, forward=True)
            #os.chdir(r'C:\Users\he1108434\Documents\EPS_test')
            #plt.title('test1')
            picName = path_file.split("\\")[-1]
            picName = picName.split(".")[0]
            print(picName)
            plt.savefig('C:\\Users\\he1108434\\Documents\\EPS_test\\' + picName)
            #plt.show()
            #break
```
